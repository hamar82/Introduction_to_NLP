```python
import string
from pymorphy2 import MorphAnalyzer
from stop_words import get_stop_words
import annoy
from gensim.models import Word2Vec, FastText
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook
```

```python
columns = ['date', 'title', 'summary', 'text', 'url']
data_df = {col: [] for col in columns}
for itm in data:
  for col in columns:
    data_df[col].append(itm[col])
data_df = pd.DataFrame(data_df, columns=columns)

data_df
```

```python
morpher = MorphAnalyzer()
sw = set(get_stop_words("ru"))
exclude = set(string.punctuation)

def preprocess_txt(line):
    spls = "".join(i for i in line.strip() if i not in exclude).split()
    spls = [morpher.parse(i.lower())[0].normal_form for i in spls]
    spls = [i for i in spls if i not in sw and i != ""]
    return spls
```

```python
def get_sentenses(corpus):
    sentences = []
    for text in tqdm_notebook(corpus):
        spls = preprocess_txt(text)
        sentences.append(spls)
    return sentences
```

Попробуем обучиться на 2-х сущностях - на заголовке статьи и на кратком содержании. Есть предположение, что лучше всего будет справляться краткое содержание - оно не слишком короткое, чтоб не уловить смысл.

Предобработка же самих тестов статей очень долгое - около 3-х часов, поэтому пробовать не буду.

```python
sent_title = get_sentenses(data_df.title)
```

```python
sent_summary = get_sentenses(data_df.summary)
```

```python
modelW2V_title = Word2Vec(sentences=sent_title, size=300, window=5, min_count=1, workers=8)
modelFT_title = FastText(sentences=sent_title, size=300, min_count=1, window=5, workers=8)
modelW2V_summary = Word2Vec(sentences=sent_summary, size=300, window=5, min_count=1, workers=8)
modelFT_summary = FastText(sentences=sent_summary, size=300, min_count=1, window=5, workers=8)
```

```python
def get_annoy_index(modelW2V, modelFT, sentenses):
    w2v_index = annoy.AnnoyIndex(300 ,'angular')
    ft_index = annoy.AnnoyIndex(300 ,'angular')
    
    counter = 0

    for sentence in tqdm_notebook(sentenses):
        n_w2v = 0
        n_ft = 0
        
        vector_w2v = np.zeros(300)
        vector_ft = np.zeros(300)
        for word in sentence:
            if word in modelW2V.wv:
                vector_w2v += modelW2V.wv[word]
                n_w2v += 1
            if word in modelFT.wv:
                vector_ft += modelFT.wv[word]
                n_ft += 1
        if n_w2v > 0:
            vector_w2v = vector_w2v / n_w2v
        if n_ft > 0:
            vector_ft = vector_ft / n_ft
        w2v_index.add_item(counter, vector_w2v)
        ft_index.add_item(counter, vector_ft)
            
        counter += 1
        
    w2v_index.build(10)
    ft_index.build(10)

    return w2v_index, ft_index
```

```python
def get_response(question, index, model, most_similar=3):
    question = preprocess_txt(question)
    vector = np.zeros(300)
    norm = 0
    for word in question:
        if word in model.wv:
            vector += model.wv[word]
            norm += 1
    if norm > 0:
        vector = vector / norm
    answers = index.get_nns_by_vector(vector, most_similar, )
    return answers
```

```python
w2v_index_title, ft_index_title = get_annoy_index(modelW2V_title, modelFT_title, sent_title)
```

```python
w2v_index_summary, ft_index_summary = get_annoy_index(modelW2V_summary, modelFT_summary, sent_summary)
```

```python
question = 'Чемпион России по футболу'
answers = get_response(question, w2v_index_title, modelW2V_title)
data_df.iloc[answers]
```

```python
answers = get_response(question, ft_index_title, modelFT_title)
data_df.iloc[answers]
```

```python
answers = get_response(question, w2v_index_summary, modelW2V_summary)
data_df.iloc[answers]
```

```python
answers = get_response(question, ft_index_summary, modelFT_summary)
data_df.iloc[answers]
```

**Выводы**:
- Поиск по заголовкам справляется хуже поиска по краткому содержанию.
- В поиске по краткому содержанию и W2V и FastText выдают хотя бы статьи на тему спорта. По заголовкам же выдаются статьи из самых разных тематик.
- Кто лучше справляется - W2V или FastText сказать трудно, в целом все они выдали статьи про спорт. Но W2V выдала одну статью именно про футбол, хотя это может быть и случайностью.
- В целом, можно сказать, что такой примитивный поиск по усреднению эмбеддингов слов, как и сама технология векторизации вполне неплохо работает.
